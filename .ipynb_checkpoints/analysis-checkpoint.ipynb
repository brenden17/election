```python
import pandas as pd
df = pd.read_csv('data/original/13.tsv', sep='\t', thousands=',')
df['result'] = df['RPR'] - df['OPR']
df['result'] = df['result'] > 0
# print(df[['local', 'result']])
```

```python
import pandas as pd

df = pd.read_csv('data/decsiontreedata.csv')
X = df[['seoul', 'incheon', 'gyeonggi', 'gangwon', 'daejeon', 'chungnam', 
            'chuugbuk', 'gwangju', 'jeonnam', 'jeonbuk', 'busan', 
            'ulsan', 'kyeongnam', 'deagu', 'kyeongbuk', 'jesu']]
y = df['result']
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC

# extraction features with decision tree
extratree = ExtraTreesClassifier()
extratree.fit(X, y)
# display the relative importance of each attribute
print(extratree.feature_importances_)

## SVM
svc = SVC()
svc.fit(X, y)
print(svc.score(X, y))

    
    
```

```python
from sklearn import tree
from sklearn.externals.six import StringIO

# extraction features with decision tree
decitiontree = tree.DecisionTreeClassifier()
decitiontree.fit(X, y)
print(decitiontree.score(X, y))

# show
with open("election.dot", 'w') as f:
    f = tree.export_graphviz(decitiontree, out_file=f)

# !dot -Tpdf election.dot -o election.pdf
!dot -Tpng election.dot > election.png

from IPython.core.display import Image 
Image(filename='election.png') 
```
